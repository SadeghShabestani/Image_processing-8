```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('pic/input/noisey_OCR.jpg',cv2.IMREAD_GRAYSCALE)

image = 255 - image

plt.imshow(image , cmap = 'gray')
```

```python
_,threshold = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
plt.imshow(threshold , cmap = 'gray')
```

```python
iamge_blur = cv2.medianBlur(threshold , 5)
plt.imshow(iamge_blur,cmap = 'gray')
```

```python
image = cv2.getRectSubPix(iamge_blur,(12,35),(43,37))
cv2.imwrite('pic/output/res0.jpg',image)
plt.imshow(image,cmap = 'gray')
```
