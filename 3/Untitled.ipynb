```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('pic\input/OCR.jpg',cv2.IMREAD_GRAYSCALE)

image = 255 - image

plt.imshow(image,cmap = 'gray')
```

```python
_,threshold = cv2.threshold(image,90,255,cv2.THRESH_BINARY)
plt.imshow(image,cmap = 'gray')
```

```python
image = cv2.getRectSubPix(threshold, (50,55), (65,82))
cv2.imwrite('pic\output/res0.jpg',image)
plt.imshow(image , cmap = 'gray')
```
